```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
% matplotlib inline
```

## FOURIER TRANSFORM

show the fourier transform of given images:

```python
''' Write a function to take an image name and print and show its fourier transform '''

def fourier(path):
    ''' Read image in grayscale '''
    image = cv.imread(path, cv.IMREAD_GRAYSCALE)
    
    print("read image")
    
    
    

    ''' Compute magnitude of fourier transform of image using numpy or opencv'''
    f_image = np.fft.fft2(image)
    f_image = np.fft.fftshift(f_image)
    f_image_magn = 20 * np.log(np.abs(f_image))
    
    print("fourier done")



    ''' plot image and magnitude using plt.imshow '''
    print("showing image")
    plt.subplot(1,2,1).axis("off")
    plt.tight_layout()
    plt.imshow(image, cmap = "gray")
    plt.title("image")
    
    plt.subplot(1,2,2).axis("off")
    plt.tight_layout()
    plt.imshow(f_image_magn, cmap = "gray")
    plt.title("fourier transform")
    plt.show()
  
```

```python
images = ['leaf.jpg', 'rocks.jpg', 'scene.jpg', 'starfish.jpg']

for i in images:
    fourier(i)
```
