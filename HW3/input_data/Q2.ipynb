```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
image = cv.imread("degraded.jpg")
clean = cv.imread("clear.jpg")
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
median = cv.medianBlur(image,3)
blur = cv.GaussianBlur(median,(5, 5),2)

```

```python
fig, axs = plt.subplots(2, 2, figsize=(15,15))
axs[0][0].axis("off")
axs[0][1].axis("off")
axs[1][0].axis("off")
axs[1][1].axis("off")

axs[0][0].set_title("degraded")
axs[0][1].set_title("after median filtering")
axs[1][0].set_title("after gaussian smoothing")
axs[1][1].set_title("the clean image")

axs[0][0].imshow(image)
axs[0][1].imshow(median)
axs[1][0].imshow(blur)
axs[1][1].imshow(clean)
plt.show()
```

The most realisable noise in the degraded picture is salt and pepper noise which can be easily removed with median filtering.
However after median filtering, I realised that there is still some noise left in the picture(especially in the sky). so I used a gaussian filter to decrease that noise. However by doing that, the details and the quality of the image decreases, but the noise will almost be removed.
