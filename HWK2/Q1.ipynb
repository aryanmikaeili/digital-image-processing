```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (25.0, 25.0)

image = cv2.imread('pic.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
bit = 8 # Between 1 and 8
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,1).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 7
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,2).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 6
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,3).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 5
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,4).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 4
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,5).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 3
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,6).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 2
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,7).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')
bit = 1
requantized_image = (image / (2**(8 - bit))).astype('uint8')
plt.subplot(1,8,8).axis("off")
plt.tight_layout()
plt.imshow(requantized_image, cmap='gray')






```

# description

As you can see in the pictures above, As we decrease the number of bits and the number of quantization layers, the pictures will get distorted. This distortion is visible from bit = 4 and on the last image which we have only two levels, we can see that the image is pure black and white.
